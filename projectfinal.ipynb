```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, adjusted_rand_score

# Read the dataset
data = pd.read_excel(r"C:\Users\91630\OneDrive\Desktop\ml\training_mathbert3solutionlatex.xlsx")

# Extract the embeddings data (383 columns)
embeddings_data = data.iloc[:, :383].values

# Extract the question numbers (384th column) and labels (385th column)
question_numbers = data.iloc[:, 384].values
labels = data.iloc[:, 385].values

# Perform t-SNE dimensionality reduction to 2 dimensions
tsne = TSNE(n_components=2, random_state=42)
embeddings_tsne = tsne.fit_transform(embeddings_data)

def evaluate_clustering(embeddings_data, ground_truth_labels, title):
    plt.figure(figsize=(15, 6))
    for k in range(2, 7):  # Considering k values 2 to 6
        kmeans = KMeans(n_clusters=k, random_state=42)
        clusters = kmeans.fit_predict(embeddings_data)

        # Evaluate clustering accuracy using adjusted Rand index
        ari = adjusted_rand_score(ground_truth_labels, clusters)

        plt.subplot(1, 5, k - 1)
        plt.scatter(embeddings_tsne[:, 0], embeddings_tsne[:, 1], c=clusters, cmap='viridis', marker='o', alpha=0.7)
        plt.title(f'Clustering with k={k}\nARI: {ari:.2f}')
        plt.xlabel('Dimension 1')
        plt.ylabel('Dimension 2')
    plt.suptitle(title, y=1.05)
    plt.tight_layout()
    plt.show()

# Evaluate clustering based on question numbers (384th column)
evaluate_clustering(embeddings_data, question_numbers, "Clustering based on Question Numbers")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances

# Read the dataset
data = pd.read_excel(r"C:\Users\91630\OneDrive\Desktop\ml\training_mathbert3solutionlatex.xlsx")

# Extract the embeddings data (383 columns)
embeddings_data = data.iloc[:, :383].values

# Perform t-SNE dimensionality reduction to 2 dimensions
tsne = TSNE(n_components=2, random_state=42)
embeddings_tsne = tsne.fit_transform(embeddings_data)

# Evaluate clustering accuracy for k=2 and k=3
k_values = [2, 3,4,5,6]
accuracy_results = []

for k in k_values:
    # Perform K-means clustering
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(embeddings_data)
    cluster_centers = kmeans.cluster_centers_
    
    # Check cluster_centers array shape and contents
    print(f"Cluster centers for k={k}:")
    print(cluster_centers)
    print(f"Shape: {cluster_centers.shape}")
    
    # Calculate pairwise distances from each point to its cluster centroid
    distances = pairwise_distances(embeddings_data, cluster_centers, metric='euclidean')
    
    cluster_accuracy = []
    
    # Calculate accuracy for each cluster
    for cluster_idx in range(k):
        # Get distances for current cluster
        cluster_distances = distances[:, cluster_idx]
        
        # Sort distances and get indices
        sorted_indices = np.argsort(cluster_distances)
        
        # Number of points in the current cluster
        num_points_in_cluster = np.sum(cluster_labels == cluster_idx)
        
        # Select top num_points_in_cluster distances
        nearest_distances_indices = sorted_indices[:num_points_in_cluster]
        
        # Count correct labels (same as cluster label)
        correct_labels_count = np.sum(cluster_labels[nearest_distances_indices] == cluster_idx)
        
        # Calculate accuracy for the current cluster
        cluster_accuracy.append(correct_labels_count / num_points_in_cluster)
    
    # Calculate overall accuracy (mean of cluster accuracies)
    overall_accuracy = np.mean(cluster_accuracy)
    accuracy_results.append(overall_accuracy)
    
    # Plot t-SNE visualization with cluster assignments and centroids
    plt.figure(figsize=(8, 6))
    plt.scatter(embeddings_tsne[:, 0], embeddings_tsne[:, 1], c=cluster_labels, cmap='viridis', s=50, alpha=0.5)
    plt.title(f"t-SNE Visualization with K-means Clustering (k={k})")
    plt.xlabel("t-SNE Component 1")
    plt.ylabel("t-SNE Component 2")
    plt.colorbar(label='Cluster Label')
    
    # Annotate cluster centroids (if cluster_centers has expected structure)
    if cluster_centers.ndim == 2 and cluster_centers.shape[1] == 2:
        for i, (center_x, center_y) in enumerate(cluster_centers):
            plt.scatter(center_x, center_y, s=200, marker='x', color='red', label=f'Centroid {i+1}')
            plt.text(center_x + 0.5, center_y + 0.5, f'Centroid {i+1}', fontsize=12)
    
    plt.legend()
    plt.show()

# Plotting accuracy results
plt.figure(figsize=(8, 6))
plt.bar(k_values, accuracy_results, color='skyblue')
plt.title('Clustering Accuracy for Different k Values')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Accuracy')
plt.ylim(0, 1.0)
plt.xticks(k_values)
plt.show()

print("Accuracy Results:")
for i, k in enumerate(k_values):
    print(f"k = {k}: Accuracy = {accuracy_results[i]:.2f}")
       
overall_accuracy_l_q = accuracy_results[k_values.index(6)]
print(f"Overall Accuracy for k=6: {overall_accuracy_l_q:.2f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import euclidean_distances

# Load training and testing data from Excel files
train_data = pd.read_excel(r"C:\Users\91630\OneDrive\Desktop\ml\training_mathbert3solutionlatex.xlsx")
test_data = pd.read_excel(r"C:\Users\91630\OneDrive\Desktop\ml\training_mathbert5additional.xlsx")

# Assuming the question labels are in the 384th column
X_train = train_data.iloc[:, :383]  # Use all columns except the last one for features
y_train = train_data.iloc[:, 384]   # Use the last column for labels

X_test = test_data.iloc[:, :383]    # Use all columns except the last one for features
y_test = test_data.iloc[:, 384]     # Use the last column for labels

# Combine training and test data for t-SNE transformation
combined_data = pd.concat([X_train, X_test])
combined_labels = pd.concat([y_train, y_test])

# Perform t-SNE on the combined data
tsne = TSNE(n_components=2, perplexity=min(30, len(combined_data)-1), random_state=42)
combined_tsne = tsne.fit_transform(combined_data)

# Separate the transformed training and test data
X_train_tsne = combined_tsne[:len(X_train)]
X_test_tsne = combined_tsne[len(X_train):]

# K-means clustering with k=6 on transformed training data
kmeans = KMeans(n_clusters=6, random_state=42)
train_cluster_labels = kmeans.fit_predict(X_train_tsne)

# Predict clusters for the transformed test data
test_cluster_labels = kmeans.predict(X_test_tsne)

# Add the resulting cluster labels to the original dataframes
train_data['Cluster'] = train_cluster_labels
test_data['Cluster'] = test_cluster_labels

# Calculate Adjusted Rand Index for training data
ari_train = adjusted_rand_score(y_train, train_cluster_labels)
print(f'Adjusted Rand Index for Training Data: {ari_train}')

# Calculate Adjusted Rand Index for test data
ari_test = adjusted_rand_score(y_test, test_cluster_labels)
print(f'Adjusted Rand Index for Test Data: {ari_test}')

# Create a scatter plot of the t-SNE results for training data
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_train_tsne[:, 0], y=X_train_tsne[:, 1], hue=train_cluster_labels, palette='viridis')
plt.title('t-SNE plot of Training Data with K-means Clusters')
plt.show()

# Plot the test data points on the same t-SNE plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_train_tsne[:, 0], y=X_train_tsne[:, 1], hue=train_cluster_labels, palette='viridis', alpha=0.5, legend='full')
sns.scatterplot(x=X_test_tsne[:, 0], y=X_test_tsne[:, 1], hue=test_cluster_labels, palette='coolwarm', marker='X', s=100, legend='full')
plt.title('t-SNE plot with Training Data Clusters and Test Data Points')
plt.show()

# Calculate accuracy based on the nearest points to the centroid for each question
def calculate_accuracy(X_train_tsne, y_train, X_test_tsne, y_test):
    accuracy_list = []
    unique_questions = y_test.unique()
    
    # Calculate centroids for each question in the training data
    centroids = {}
    for question in y_train.unique():
        question_points = X_train_tsne[y_train == question]
        centroids[question] = np.mean(question_points, axis=0)
    
    for question in unique_questions:
        if question not in centroids:
            continue
        
        # Get the points corresponding to the current question in the test data
        question_points_test = X_test_tsne[y_test == question]
        n = len(question_points_test)
        
        if n == 0:
            continue
        
        # Get the centroid for the current question from the training data
        centroid = centroids[question]
        
        # Calculate distances from the centroid to all test points
        distances = euclidean_distances([centroid], X_test_tsne)[0]
        
        # Get the indices of the n nearest points
        nearest_indices = np.argsort(distances)[:n]
        
        # Print the distances for the nearest points
        nearest_distances = distances[nearest_indices]
        print(f"Question: {question}, Nearest distances: {nearest_distances}")
        
        # Check how many of the nearest points belong to the same question
        correct_labels = y_test.iloc[nearest_indices]
        correct_count = sum(correct_labels == question)
        
        # Calculate accuracy for the current question
        accuracy = correct_count / n
        accuracy_list.append(accuracy)
    
    # Calculate and return the overall accuracy
    overall_accuracy = np.mean(accuracy_list)
    return overall_accuracy

# Calculate and print the accuracy
overall_accuracy_l_q1= calculate_accuracy(X_train_tsne, y_train, X_test_tsne, y_test)
print(f'Overall accuracy based on nearest points to centroid: {overall_accuracy_l_q1}')
```
